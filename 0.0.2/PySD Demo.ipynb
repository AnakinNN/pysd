#PySD 0.0.2 Demo


### New Functionality
- This version adds the ability to modify model parameters (or for that matter, set any aux or flow to a constant) and execute the model without losing the original values of the parameters.
- inspect the parameters of the model which are not dependent on other variables, and are not stocks. These are what we are most likely to be modifying.

### New Structure
- This version brings the execution network into the PySD main file, as it will be common to all models, regardless of if they are built by reading XMILE, or via some other method (possibly vensim text output).
- There is the beginning of a set of unit tests to aid in further development
- Better inline documentation and commenting

### Limitations
v0.0.2 parses a subset of the SMILE language, including:

    "exprnd" / "exp" / "sin" / "cos" / "abs" / "int" / "inf" / "log10" / "pi" /
    "sqrt" / "tan" / "lognormal" / "normal" / "poisson" / "ln" / "min" / "max" /
    "random" / "arccos" / "arcsin" / "arctan" / "if_then_else"

But excluding any time-dependant functions (ie `"pulse"`) and any non-markov functions (ie `"delay"`).

We can also set the parameters in the model

```python
output = model.run_pandas(params={'normal_backlog':10, 'normal_delivery_rate':10})
output.head()
```

With these new features we can begin to do interesting things, such as sweep over parameter values. Here we see how changes in the initial backlog influence the backlog at the end of the simulation.

```python
test_backlog = pd.DataFrame(data=range(500, 10000, 500), columns=['normal_backlog'])
test_backlog['t100_backlog'] = test_backlog['normal_backlog'].apply(lambda x: model.run_pandas({'normal_backlog':x})['backlog'].iloc[-1])
plt.plot(test_backlog['normal_backlog'], test_backlog['t100_backlog'], '-o');
```

We can also run optimizations - say we want the final value of the backlog (at t100) to be 1000. (For obscure reasons only known to management.) What should we set our 'normal_backlog' parameter to?

```python
import scipy.optimize
# run the optimization, minimizing the square difference between the last model output and the value 1000.
res = scipy.optimize.fmin(lambda x: (model.run_pandas({'normal_backlog':x})['backlog'].iloc[-1] - 1000)**2, x0=8000)
print 'set normal backlog to', res[0]
model.run_pandas({'normal_backlog':res[0]}).plot();
```
