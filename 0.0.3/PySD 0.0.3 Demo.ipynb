#pysd 0.0.3: basic vensim functionality

#### Updates in version 0.0.3

- lowercase class name (capitalizing PySD was getting annoying)
- ability to load vensim model files


```python
import pysd
import pandas as pd
```

##Load XMILE file

We've got a parser for xmile files that currently pulls in the model structural components of the xmile file, and in the future may pull in the display components as well. For now we are limited to a subset of the full XMILE functionality.

```python
cg_model = pysd.read_XMILE('corporate-growth.xml')
```

## Inspect Python version of the model
pysd translates the xmile model equations into a python-native function which represents the derivative of the system given a particular set of stock values and time `t`.

We can inspect the function directly:

```python
print cg_model.get_derivative_function_string()
```

##View initial conditions and simulation parameters
The direct string form output of the model itself is a list of initial conditions and execution parameters

```python
print cg_model
```

## View model 'free parameters'
We can also view the 'free parameters' of the model - those whose modification does not change the model's structure.

```python
cg_model.get_free_parameters()
```

## Run the model
Running the model is straightforward.

```python
output1 = cg_model.run()
output1.plot() #this functionality provided by pandas
output1.head()
```

## Run with modified model parameters
To modify parameters before running the model, simply include a dictionary of parameter names and new values. These get applied to a copy of the dependency tree, and the original values are kept.

```python
output2 = cg_model.run(params={'normal_sales_effectiveness':360})
output2['backlog'].plot()
output1['backlog'].plot(title='Modified parameters in green');
```

## Return a numpy array instead of a pandas dataframe
If we so desire - ie, we're putting results into Scikit-learn or another package which is numpy-based - we can return just a numpy vector.

```python
cg_model.run(return_type='numpy')
```

#Load vensim file
We can also load vensim models, which are translated to the same internal representation as XMILE files, and then provide the same functionality.

```python
bank_model = pysd.read_vensim('bank_balance.mdl')
print bank_model.get_free_parameters()
```

#Bonus: the pysd execution network

pysd stores all of its equations and dependencies (and in future releases additional parameters for each component like units, limit values, etc) in a directed network. This ensures that all of the components are executed in the right sequence, and will let us 'amputate' parts of the model if they are zeroed out by the user.

We can plot the execution network to check how pysd interpreted those dependencies. Sometimes this is useful, often it isnt.

This is a bit of an easter egg - as its a private function of the class, and isnt really very well developed. 

```python
bank_model._draw_model_network()
```

```python
cg_model._draw_model_network()
```
