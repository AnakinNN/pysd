```python
%pylab inline
import pandas as pd
import json

with open('speedtest_results.json') as f:
    #data = pd.DataFrame(json.loads(line) for line in f)
    data = pd.DataFrame([eval(line) for line in f])
    #data.set_index('timestamp', inplace=True)
    data.drop('version', axis=1, inplace=True)
    
data /= data.max() 
data.plot(subplots=True, figsize=(10,10), ylim=(0,1));
```
